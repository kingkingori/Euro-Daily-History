# Daily Exchange Rates per Euro 1999-
time series of foreign exchange rates per 1 Euro

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import datetime as dt
```

```python
euro_hist = pd.read_csv('euro-daily-hist_1999_2022.csv')
euro_hist
```

```python
euro_hist_us = euro_hist.loc[:, ['Period\\Unit:', '[US dollar ]']]
euro_hist_us
```

```python
euro_hist_us.info()
```

```python
# accomodate special characters, escape when needed
euro_hist_us.rename(columns={'Period\\Unit:': 'date', '[US dollar ]': 'usd'}, inplace=True)
```

```python
euro_hist_us
```

```python
euro_hist_us.info()
```

```python
euro_hist_us['date'] = pd.to_datetime(euro_hist_us['date'])
```

```python
# an entry with value '-', coerce it to 'NaN'
euro_hist_us['usd'] = pd.to_numeric(euro_hist_us['usd'], errors='coerce')
euro_hist_us['usd'].dtype
```

```python
euro_hist_us
```

```python
plt.plot(euro_hist_us['date'], euro_hist_us['usd'].rolling(1).mean())
plt.show()
```

#### There are gaps in the chart due to nulls in the data

```python
# remove nulls causing gaps in the chart -- dropna()
euro_hist_us = euro_hist_us[~euro_hist_us['usd'].isna()]
```

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(16, 8))

fig.suptitle('EUR vs USD', fontsize=16, weight='bold')

"""
axs[0, 0].plot(euro_hist_us['date'], euro_hist_us['usd'].rolling(1).mean())
axs[0, 1].plot(euro_hist_us['date'], euro_hist_us['usd'].rolling(10).mean())
axs[1, 0].plot(euro_hist_us['date'], euro_hist_us['usd'].rolling(50).mean())
axs[1, 1].plot(euro_hist_us['date'], euro_hist_us['usd'].rolling(100).mean())
axs[2, 0].plot(euro_hist_us['date'], euro_hist_us['usd'].rolling(200).mean())
axs[2, 1].plot(euro_hist_us['date'], euro_hist_us['usd'].rolling(365).mean())
"""
"""the above codeblock may be replaced as below"""

axs_list = [axs[0, 0], axs[0, 1], axs[1, 0], axs[1, 1], axs[2, 0], axs[2, 1]]
periods = [1, 10, 50, 100, 200, 365]

for ax, period in zip(axs_list, periods):
    ax.plot(euro_hist_us['date'], euro_hist_us['usd'].rolling(period).mean())
    ax.set_title(f'Rolling Window: {period}', weight='bold')

    ax.set_ylim([0.8, 1.7])
    ax.set_xlim([dt.datetime(1998, 1, 1), dt.datetime(2024, 12, 31)])

    """per year"""
    # ax.axvline([dt.datetime(yr, 1, 1) for yr in range(2000, 2024, 4)], alpha=0.2)
    ax.axvline(dt.datetime(2000, 1, 1), alpha=0.2)
    ax.axvline(dt.datetime(2004, 1, 1), alpha=0.2)
    ax.axvline(dt.datetime(2008, 1, 1), alpha=0.2)
    ax.axvline(dt.datetime(2012, 1, 1), alpha=0.2)
    ax.axvline(dt.datetime(2016, 1, 1), alpha=0.2)
    ax.axvline(dt.datetime(2020, 1, 1), alpha=0.2)
    ax.axvline(dt.datetime(2024, 1, 1), alpha=0.2)

    """the housing crisis of '08 and COVID"""
    ax.axvspan(dt.datetime(2007, 1, 1), dt.datetime(2008, 12, 31), alpha=0.3)
    ax.axvspan(dt.datetime(2020, 1, 1), dt.datetime(2023, 5, 1), alpha=0.3)
    
    """Bush, Obama, and Trump presidencies"""
    # ax.axvspan(dt.datetime(2001, 1, 1), dt.datetime(2009, 1, 1), color='c', ec='y', alpha=0.3)
    # ax.axvspan(dt.datetime(2009, 1, 1), dt.datetime(2017, 1, 1), color='c', ec='y', alpha=0.3)
    # ax.axvspan(dt.datetime(2017, 1, 1), dt.datetime(2021, 1, 1), color='c', ec='y', alpha=0.3)

# for ax in axs_list:
#     ax.axvline([dt.datetime(yr, 1, 1) for yr in range(2000, 2024, 4)], alpha=0.2)
    
plt.tight_layout()
plt.show()
```
